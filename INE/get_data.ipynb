```python
with open('62490.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python

clean_name('Total Nacional, Viajero, Total')
```

```python
df_viajeros = pd.DataFrame(columns=['city','Agno','Mes','Value'])
df_pernoctaciones = pd.DataFrame(columns=['city','Agno','Mes','Value'])
```

```python
counter = 0
for i in data:
    if i['Nombre'].split(',')[-1] == ' Total':
        if i['Nombre'].split(',')[-2] == ' Pernoctaciones':
            for data_data in i['Data']:
                Agno = data_data['NombrePeriodo'].split('M')[0]
                Mes = data_data['NombrePeriodo'].split('M')[1]
                city = clean_name(i['Nombre'])
                df_pernoctaciones.loc[counter] = [normalizar_nombre(city), Agno, Mes, data_data['Valor']]
                counter += 1
        elif i['Nombre'].split(',')[-2] == ' Viajero':
            for data_data in i['Data']:
                Agno = data_data['NombrePeriodo'].split('M')[0]
                Mes = data_data['NombrePeriodo'].split('M')[1]
                city = clean_name(i['Nombre'])
                df_viajeros.loc[counter] = [normalizar_nombre(city), Agno, Mes, data_data['Valor']]
                counter += 1
```
