```python
pernoctaciones = pd.read_csv('pernoctaciones.csv')
viajeros = pd.read_csv('viajeros.csv')
precios_hoteleros = pd.read_csv('precios_hoteleros.csv')
ipc_restaurantes_hoteles = pd.read_csv('IPC/IPC_restaurantes_hoteles.csv')
precios_hoteleros
```

```python
df_merged = pd.merge(pernoctaciones, viajeros, on=['city', 'Agno', 'Mes'], how='inner')
df_merged.rename(columns={'Value_x': 'pernoctaciones', 'Value_y': 'viajeros', 'Value_z': 'Value_precios'}, inplace=True)
df_merged = pd.merge(df_merged, precios_hoteleros, on=['city', 'Agno', 'Mes'], how='inner')
df_merged = pd.merge(df_merged, ipc_restaurantes_hoteles, on=['city', 'Agno', 'Mes'], how='inner')
df_merged.rename(columns={'Value_x': 'precios_hoteleros', 'Value_y': 'ipc_restaurantes_hoteles', 'Value_z': 'Value_precios'}, inplace=True)
df_merged
```

```python
df_final = df_merged[(df_merged['city'] != 19) & (df_merged['city'] != 18)]
df_final['Date'] = pd.to_datetime(df_final['Agno'].astype(str) + '-' + df_final['Mes'].astype(str), format='%Y-%m')
df_final.drop(columns=['Agno', 'Mes'], inplace=True)
df_final.to_csv('INE_data.csv', index=False)
```
